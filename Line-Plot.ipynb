# Line Pots (Series/Dataframe) <a id="4"></a>


**Let's start with a case study:**

In 2010, Haiti suffered a catastrophic magnitude 7.0 earthquake. The quake caused widespread devastation and loss of life and about three million people were affected by this natural disaster. As part of Canada's humanitarian effort, the Government of Canada stepped up its effort in accepting refugees from Haiti. We can quickly visualize this effort using a `Line` plot:

**Question:** Plot a line graph of immigration from Haiti using `df.plot()`.


First, we will extract the data series for Haiti.


````python
#Since we converted the years to string, 
#let's declare a variable that will allow us to easily call upon the full range of years:
years = list(map(str, range(1980, 2014)))
#creating data series
haiti = df_can.loc['Haiti', years] # passing in years 1980 - 2013 to exclude the 'total' column
haiti.head()
````

Next, we will plot a line plot by appending `.plot()` to the `haiti` dataframe.


````python
haiti.index = haiti.index.map(int) 
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

# annotate the 2010 Earthquake. 
# syntax: plt.text(x, y, label)
plt.text(2000, 6000, '2010 Earthquake') # see note below

plt.show() 
````

We can clearly notice how number of immigrants from Haiti spiked up from 2010 as Canada stepped up its efforts to accept refugees from Haiti.

With just a few lines of code, you were able to quickly identify and visualize the spike in immigration!

Quick note on x and y values in `plt.text(x, y, label)`:
    
     Since the x-axis (years) is type 'integer', we specified x as a year. The y axis (number of immigrants) is type 'integer', so we can just specify the value y = 6000.
    
```python
    plt.text(2000, 6000, '2010 Earthquake') # years stored as type int
```
    If the years were stored as type 'string', we would need to specify x as the index position of the year. Eg 20th index is year 2000 since it is the 20th year with a base year of 1980.
```python
    plt.text(20, 6000, '2010 Earthquake') # years stored as type int
```
    We will cover advanced annotation methods in later modules.


We can easily add more countries to line plot to make meaningful comparisons immigration from different countries. 

**Question:** Let's compare the number of immigrants from India and China from 1980 to 2013.


Step 1: Get the data set for China and India, and display the dataframe.


````python


df_CI = df_can.loc[['India', 'China'], years]
df_CI
````

Step 2: Plot graph. We will explicitly specify line plot by passing in `kind` parameter to `plot()`.


````python
df_CI = df_CI.transpose()
df_CI.head()
````

````python

df_CI.index = df_CI.index.map(int) # let's change the index values of df_CI to type integer for plotting
df_CI.plot(kind='line')

plt.title('Immigrants from China and India')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
````

<br>From the above plot, we can observe that the China and India have very similar immigration trends through the years. 


*Note*: How come we didn't need to transpose Haiti's dataframe before plotting (like we did for df_CI)?

**Question:** Compare the trend of top 5 countries that contributed the most to immigration to Canada.


````python


 #Step 1: Get the dataset. Recall that we created a Total column that calculates cumulative immigration by country. 
#We will sort on this column to get our top 5 countries using pandas sort_values() method.
    
inplace = True # paramemter saves the changes to the original df_can dataframe
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)

# get the top 5 entries
df_top5 = df_can.head(5)

# transpose the dataframe
df_top5 = df_top5[years].transpose() 

print(df_top5)


#Step 2: Plot the dataframe. To make the plot more readeable, we will change the size using the `figsize` parameter.
df_top5.index = df_top5.index.map(int) # let's change the index values of df_top5 to type integer for plotting
df_top5.plot(kind='line', figsize=(14, 8)) # pass a tuple (x, y) size



plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')


plt.show()
````

### Other Plots

 There are many other plotting styles available other than the default Line plot, all of which can be accessed by passing `kind` keyword to `plot()`. The full list of available plots are as follows:

* `bar` for vertical bar plots
* `barh` for horizontal bar plots
* `hist` for histogram
* `box` for boxplot
* `kde` or `density` for density plots
* `area` for area plots
* `pie` for pie plots
* `scatter` for scatter plots
* `hexbin` for hexbin plot


# Summary of the Data Analysis on Immigration Trends to Canada



This analysis examines immigration trends to Canada, focusing on the fluctuations in immigration numbers over time and identifying key factors contributing to these trends. The dataset used in this study contains detailed records of immigration from various countries to Canada from 1980 to 2013. The goal of the analysis is to visualize and interpret the immigration data for specific countries, compare trends, and identify any significant events that may have impacted immigration rates.

## Data Overview
The dataset consists of immigration data from 195 countries, with each country's data spanning from 1980 to 2013. The dataset is structured in a way that allows for easy filtering and aggregation of immigration data by country and year. The data includes the number of immigrants from each country for each year, with a total column that sums up the immigration numbers across all years for each country.

## Exploratory Data Analysis (EDA)
The initial step in the analysis was to explore the dataset and understand the structure of the data. We cleaned and organized the data by sorting it based on the total number of immigrants from each country, ensuring that the most relevant countries could be easily identified. The years were treated as integers for easier plotting and visualization. Various countries were examined individually to spot any trends or irregularities, and it was observed that some countries exhibited consistent immigration patterns, while others showed spikes in specific years.

## Immigration Trend of Haiti
Haiti was one of the countries examined in detail. The analysis showed that the number of immigrants from Haiti to Canada increased significantly after the 2010 earthquake, which prompted Canada to step up its efforts to accept refugees from Haiti. This event marked a clear spike in immigration numbers, which was easily identifiable through a line chart. The annotation of the 2010 earthquake was included in the plot to highlight this significant event.
The code analysis also pointed out that the data for Haiti was a series rather than a dataframe, which made it easier to plot without needing transposition, unlike dataframes with multiple countries.

## Comparison of Immigrants from India and China
A comparison was made between the immigration trends of two countries—India and China—over the same period (1980-2013). By filtering the dataset for both countries and plotting their immigration numbers, it was observed that both countries followed similar trends, with consistent immigration flows throughout the years. The line chart revealed that while there were fluctuations, the general immigration numbers remained relatively stable for both countries.
This comparison helped underline the importance of understanding immigration patterns from key regions and offered insights into how certain countries may have been impacted by economic or geopolitical factors influencing immigration to Canada.

## Top 5 Contributing Countries
The next analysis focused on comparing the immigration trends of the top 5 countries contributing the most immigrants to Canada over the years. By sorting the dataset based on the total immigration numbers and selecting the top 5 countries, the trends for India, China, the United Kingdom, the Philippines, and Pakistan were plotted.
These countries showed varying trends in immigration, with some countries, like India and China, maintaining relatively consistent numbers over the years, while others, like the Philippines and Pakistan, saw more significant fluctuations. This variation can be attributed to numerous factors, such as political events, international relations, and Canadian immigration policies over the years.
The analysis of these top 5 countries revealed that the United Kingdom was a significant contributor, especially in the earlier years, while countries like the Philippines and Pakistan showed noticeable increases in immigration numbers during certain periods. The line chart of the top 5 countries provided a clear visualization of these trends, making it easier to compare the contributions of each country over the 33-year period.

## Key Observations
Several key observations were made during the analysis:

Spikes in Immigration: Certain years showed significant spikes in immigration numbers, which could be linked to specific events or changes in Canadian immigration policies.
Geopolitical Factors: The fluctuations in immigration numbers were not random but were likely influenced by geopolitical events in the source countries. For example, the spike in immigration from Haiti following the 2010 earthquake is a clear example of how global events affect immigration patterns.
Consistent Contributors: Countries like India and China consistently contributed large numbers of immigrants to Canada over the years, which could reflect stable immigration policies and ongoing demand for immigration from these countries.
Shifts in Top Countries: The ranking of the top contributing countries changed over time, with countries like the Philippines and Pakistan gaining prominence in later years. This shift could be attributed to both internal factors within these countries and changes in Canadian immigration policy that targeted specific regions.
Conclusion
This detailed analysis of immigration trends to Canada provides valuable insights into the factors that drive immigration flows from different countries. By visualizing the data, it became clear how significant events, such as the 2010 earthquake in Haiti, and long-term trends, like the consistent immigration from India and China, have shaped the overall picture of Canadian immigration. The comparison of the top 5 countries contributing the most immigrants to Canada highlighted the evolving nature of global migration patterns and the role of Canadian immigration policies in shaping these trends.
The analysis also demonstrated the power of data visualization in making complex trends more accessible and understandable. With just a few lines of code, meaningful insights can be drawn from large datasets, allowing policymakers, researchers, and analysts to make informed decisions based on historical trends.
